```python
import pandas as pd
```

```python
df=pd.read_csv('Housing_Prices.csv')
```

```python
df
```

```python
Y=df[['Sale Price']]
```

```python
X=df.drop(['Date House was Sold','Sale Price'] ,axis='columns')
```

```python
X
```

```python
Y.mean()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
```

```python
X['Waterfront View']=le1.fit_transform(X['Waterfront View'])
```

```python
le2=LabelEncoder()
X['Condition of the House']=le2.fit_transform(X['Condition of the House'])
```

```python
le3=LabelEncoder()
X['No of Times Visited']=le3.fit_transform(X['No of Times Visited'])
```

```python
X
```

```python
X=X.drop(['ID'],axis='columns')
```

```python
X=X.drop(['Lot Area (in Sqft)','Area of the House from Basement (in Sqft)','Basement Area (in Sqft)','Renovated Year'] ,axis='columns')
```

```python
X
```

```python
X.isna().sum()
```

```python
X['Flat Area (in Sqft)']=X['Flat Area (in Sqft)'].fillna(X['Flat Area (in Sqft)'].mean())
```

```python
X['No of Bathrooms']=X['No of Bathrooms'].fillna(X['No of Bathrooms'].mean())
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
```

```python
X_scaled=s.fit_transform(X)
```

```python
Y_scaled=s.fit_transform(Y)
```

```python
Y=Y.fillna(Y.mean())
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
from sklearn.model_selection import train_test_split
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```
